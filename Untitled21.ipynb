```python
import pandas as pd
file_path = '/content/drive/MyDrive/datasettt/datset.xlsx'
# Use pd.read_excel to read Excel files
data = pd.read_excel(file_path)
```

```python
print("Dataset Shape:", data.shape)
print("\nColumn Names:", data.columns)
print("\nData Types:\n", data.dtypes)
print("\nFirst 5 Rows:\n", data.head())
```

```python
print("\nMissing Values:\n", data.isnull().sum())
```

```python
print("\nSummary Statistics (Numerical Features):\n", data.describe())
print("\nSummary Statistics (Categorical Features):\n", data.describe(include='object'))
```

```python
data['Total Spend'] = pd.to_numeric(data['Total Spend'], errors='coerce')
data['Payment Delay'] = pd.to_numeric(data['Payment Delay'], errors='coerce')
```

```python
categorical_features = ['Gender', 'Subscription Type', 'Contract Length', 'Churn']
for col in categorical_features:
    data[col] = data[col].astype('category')
```

```python
fig, axes = plt.subplots(1, len(numerical_features), figsize=(5 * len(numerical_features), 4))

for i, col in enumerate(numerical_features):
    sns.histplot(data[col], kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, len(categorical_features), figsize=(5 * len(categorical_features), 6))

for i, col in enumerate(categorical_features):
    sns.countplot(y=data[col], ax=axes[i])
    axes[i].set_title(f'Count of {col}')
    axes[i].set_ylabel("")  # Optionally remove y-axis labels to avoid overlap
    axes[i].set_xlabel("Count")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Churn')
plt.title("Churn Count")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Subscription Type', hue='Churn')
plt.title("Churn by Subscription Type")
plt.show()
```

```python
encoded_data = data.copy()
encoded_data['Gender'] = encoded_data['Gender'].cat.codes
encoded_data['Subscription Type'] = encoded_data['Subscription Type'].cat.codes
encoded_data['Contract Length'] = encoded_data['Contract Length'].cat.codes
encoded_data['Churn'] = encoded_data['Churn'].cat.codes
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = encoded_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Subscription Type', y='Total Spend', hue='Churn')
plt.title("Total Spend by Subscription Type and Churn")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.boxplot(data=data, x='Churn', y='Support Calls', ax=axes[0])
axes[0].set_title("Support Calls by Churn")

sns.boxplot(data=data, x='Churn', y='Payment Delay', ax=axes[1])
axes[1].set_title("Payment Delay by Churn")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Subscription Type', y='Age')
plt.title("Age Distribution by Subscription Type")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Contract Length', y='Last Interaction', hue='Churn')
plt.title("Last Interaction by Contract Length and Churn")
plt.show()
```
